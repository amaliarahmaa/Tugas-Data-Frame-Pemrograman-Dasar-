```python
import pandas as pd 

tgs_excel = pd.read_excel('datasampah.xlsx', sheet_name='data')
tgs_excel
```

```python
# Soal 1

df_produksi_sampah = tgs_excel[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']]
print(df_produksi_sampah)
```

```python
# Soal 2

total_produksi_sampah = 0
tahun_produksi = int(input("masukkan tahun berapa yang ingin di cek total produksi sampah"))

while (tahun_produksi < 2015 or tahun_produksi > 2023):
    tahun_produksi = int(input("masukkan tahunnya untuk cek produksi sampah diantara tahun 2015 sampai 2023"))

for index, baris in df_produksi_sampah.iterrows():
    if baris['tahun'] == tahun_produksi:
        jumlah_produksi_sampah = total_produksi_sampah + baris['jumlah_produksi_sampah']
        total_produksi_sampah = jumlah_produksi_sampah

print (f"jumlah sampah berdasarkan tahun yang anda cek adalah : {total_produksi_sampah} ton")
```

```python
# Soal 3

data_tahun = {}

for index, baris in df_produksi_sampah.iterrows():
    data_tahun[baris['tahun']] = 0

for i in data_tahun.keys():
    for index, baris in df_produksi_sampah.iterrows():
        if baris['tahun'] == int(i):
            data_tahun[i] = data_tahun[i] + baris['jumlah_produksi_sampah']

for i, j in data_tahun.items():
    print(f"tahun {i} : jumlah sampah per ton nya yaitu {j}")
```

```python
# soal 4
data_kota_kab = {}
for index, baris in df_produksi_sampah.iterrows():

    if baris['tahun'] not in data_kota_kab:
        data_kota_kab[baris['tahun']] = {}

    if baris['nama_kabupaten_kota'] not in data_kota_kab[baris['tahun']]:
        data_kota_kab[baris['tahun']][baris['nama_kabupaten_kota']] = 0
    

    data_kota_kab[baris['tahun']][baris['nama_kabupaten_kota']] += baris['jumlah_produksi_sampah']

for i, j in data_kota_kab.items():
    for a, b in j.items():
        print(f"untuk tahun {i}, dari kota/kab {a}, jumlah sampah per ton yaitu {b}")
```

```python
# Export excel

df_produksi_sampah.to_excel('tugas_dataframe.xlsx', index= False)
```

```python
# Export csv

df_produksi_sampah.to_csv('tugas_dataframe.csv', index= False)
```
